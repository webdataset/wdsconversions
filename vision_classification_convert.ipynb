```python
fname = "mnist"
DATASET = datasets.MNIST
ds = DATASET(root="./__data__", download=True, train=True)
with wds.TarWriter(f"./shards/{fname}-train.tar") as ostream:
    for i, (img, cls) in enumerate(ds):
        sample = dict(
            __key__=str(i),
            png=img,
            cls=cls,
        )
        ostream.write(sample)
ds = DATASET(root="./__data__", download=True, train=False)
with wds.TarWriter(f"./shards/{fname}-test.tar") as ostream:
    for i, (img, cls) in enumerate(ds):
        sample = dict(
            __key__=str(i),
            png=img,
            cls=cls,
        )
        ostream.write(sample)
!ls -lt shards | head
```
